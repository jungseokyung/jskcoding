```python
def date_format1(d):
    d = str(d).replace('-','.')
    yyyy = int(d.split('.')[0])
    mm = int(d.split('.')[1])
    dd = int(d.split('.')[2])
    
    this_date = dt.date(yyyy,mm,dd)

    return this_date
```

```python
def date_format2(d):
    yyyy = int(d[:4])
    mm = int(d[4:6])
    dd = int(d[6:])
    
    this_date = dt.date(yyyy,mm,dd)

    return this_date
```
