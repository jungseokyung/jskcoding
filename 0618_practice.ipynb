```python
x=np.arange(-5,5,1)
y1=0.3*(x-5)**2+1
y2=-1.5*x+3
y3=np.sin(x)**2
y4=10*np.exp(-x)+1

#2X2 행렬로 나누고 위치조정

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

```python
import  numpy as np

#데이터 생성
x=np.arange(0,10,0.1)
y1=0.3*(x-5)**2+1
y2=-1.5*x+3
y3=np.sin(x)**2 #Nump에서 sin은 np.sin()으로 입력
y4=10*np.exp(-x)+1   #numpy 에서 exp는 np.exp()로 입력

#2x2 행렬이 이뤄진 하위 그래프에서 p에 다라 위치를 지정

plt.subplot(2,2,1) #p =1
plt.plot(x,y1)

plt.subplot(2,2,2) #p =2
plt.plot(x,y2)

plt.subplot(2,2,3) #p =3
plt.plot(x,y3)

plt.subplot(2,2,4) #p=4
plt.plot(x,y4)

plt.show()
```

###### LinearRegression_model_comprehension

```python
from sklearn.datasets import make_regression #가상의 회귀분석용 데이터 생성
from sklearn.linear_model import LinearRegression # LinearRegression
from sklearn.model_selection import train_test_split #Trainset,Testset 분리
```

```python
X,y=make_regression(n_samples=10000,n_features=1,bias=10,noise=0.1)
print('input_feature: \n {}\nTarget_value: \n'.format(X[:5],y[:5]))
```

```python
print(type(X),type(y))
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],y)
plt.show()
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2,shuffle=True)
```

```python
model=LinearRegression() #fit_intercept = 상수항 유무 결정

#fit 메서드로 모형 추정 (오그멘테이션은 자동으로 수행)
model_lr=model.fit(X_train,y_train)

print('weight: ', model_lr.coef_)
print('bias: ', model_lr.intercept_)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
