```python
def int_to_date_format(d):

    import datetime

    yyyy = int(d[0:4])

    mm = int(d[4:6])

    dd = int(d[6:8])

    this_date = datetime.date(yyyy, mm, dd)

    return this_date
```

```python
def date_dot_format(d):

    import datetime

#   d = str(d).replace('-', '.')


    yyyy = int(d.split('.')[0])

    mm = int(d.split('.')[1])

    dd = int(d.split('.')[2])


    this_date = datetime.date(yyyy, mm, dd)

    return this_date
```
